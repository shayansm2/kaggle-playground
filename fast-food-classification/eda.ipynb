```python
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline

import tensorflow as tf
from tensorflow import keras

from keras.preprocessing.image import load_img
```

```python
def get_all_files_name_in_path(path: str):
    return [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]


def show_the_first_image_in_path(path: str):
    name = get_all_files_name_in_path(path)[1]
    return f'{path}/{name}'
    
    
def get_path_of_food(name:str):
    return f'./data/Train/{name}'
```

```python
load_img(show_the_first_image_in_path(get_path_of_food('Baked Potato')))
```
