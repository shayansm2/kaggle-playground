```python
import pandas as pd
import seaborn as sns
```

```python
def get_info_about_dataset(dataset_name: str):
    df_train = pd.read_csv(f'./data/emnist/emnist-{dataset_name}-train.csv', header=None)
    mapping = dict()
    with open(f'./data/emnist/emnist-{dataset_name}-mapping.txt', 'r') as file:
        line = file.readline()
        while line:
            data = line.split("\n")[0].split(" ")
            key, val = data[0], data[1]
            mapping[int(key)] = chr(int(val))
            line = file.readline()
    print('chars:', ','.join(list(map(lambda x:mapping[x], df_train[0].unique().tolist()))))
    print('count of chars:', len(mapping))
    sns.histplot(df_train[0])
```

```python
get_info_about_dataset('byclass')
```

```python
get_info_about_dataset('balanced')
```

```python
get_info_about_dataset('bymerge')
```
